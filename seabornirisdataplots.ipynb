```python
data = sns.load_dataset('iris')
data.head()
```

```python
### line plot

import warnings
warnings.filterwarnings('ignore')
sns.lineplot(x='sepal_length',y='sepal_width',data=data,hue='species')
```

# Scatter Plot

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=data,hue='species')
```

```python
sns.scatterplot(x='petal_length',y='petal_width',data=data,hue='species',markers=False)
```

```python
sns.barplot(x='species', y='sepal_length',data=data,hue='species')
```

# Box Plot
- Find outliers
- divides data into 4 quartile


```python
sns.boxplot(x='sepal_length',data=data,legend='auto')
```

```python
sns.boxplot(x='sepal_width',data=data,legend='auto')
```

```python
import numpy as np
data.loc[data['sepal_width']>4,'sepal_width']=np.median(data['sepal_width'])
```

```python
sns.boxplot(y='sepal_width',data=data,hue='species')
```

# violin plot
